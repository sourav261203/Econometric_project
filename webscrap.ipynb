```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url="https://www.ambitionbox.com/list-of-companies?campaign=homepage_companies_widget&page=1"
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
web=requests.get(url,headers=headers).text
```

```python
soup=BeautifulSoup(web,'lxml')
```

```python
for i in soup.find_all('h2'):
  print(i.text.strip())
```

```python
len(soup.find_all('a',class_='review-count'))
```

```python
company=soup.find_all('div',class_='company-content-wrapper')
len(company)
```

```python
name=[]
rating=[]
review=[]
ctype=[]
hquater=[]
how_old=[]
employee=[]
for i in company:
  name.append(i.find('h2').text.strip())
  rating.append(i.find('p',class_='rating').text.strip())
  review.append(i.find('a',class_='review-count').text.strip())
  ctype.append(i.find_all('p', class_='infoEntity')[0].text.strip())
  hquater.append(i.find_all('p', class_='infoEntity')[1].text.strip())
  how_old.append(i.find_all('p', class_='infoEntity')[2].text.strip())
  try:
    employee.append(i.find_all('p', class_='infoEntity')[3].text.strip())
  except:
    employee.append(np.nan)
dict={'name': name, 'rating':rating,"review":review,"company_type":ctype,"head_quater":hquater,"how_old":how_old,"employee":employee}
```

```python
df=pd.DataFrame(dict)
df
```
